```python
from email.parser import BytesParser
from itertools import chain
from email.utils import getaddresses
```

```python
msg = b'''
From: wayne@example.com
To: Person Man <person.man@example.com>, Fredrick Douglas <music.man@example.org>
Cc: Guido <bdfl@example.com>, FLUFL <barry@example.com>
Bcc: spanish.inquisition@example.com, The Dude <big.lebowski@example.net>
Subject: Testing email

This isn't a very fancy email, but I'm just trying to prove a point here, OK?
'''.strip()
msg1 = b'''
From: wayne@example.com
To: 
Cc: Guido <bdfl@example.com>, FLUFL <barry@example.com>
Bcc: spanish.inquisition@example.com, The Dude <big.lebowski@example.net>
Subject: Testing email

This isn't a very fancy email, but I'm just trying to prove a point here, OK?
'''.strip()
email = BytesParser().parsebytes(msg1)
```

```python
email_list = []
for recipient in getaddresses(chain(email.get_all('to', []), email.get_all('cc', []), email.get_all('bcc', []))):
    #print(f"receipient: {recipient[1]}")
    email_list.append(recipient[1])
    
```

```python

gmail_user = "sandeepduttacse45@gmail.com"
SUBJECT = "Interview Tips from Sandeep Dutta"
message= '''
Hi friends hope you are doing well.
So here is my below some interview tips which will help you to crack the interview.
So here is the link for it which is a 1 to 2 min read
Tips: https://link.medium.com/vjrNOVcL03 

do share and clap after reading.

Also all the best for your interviews.
'''
```

```python
for i in range(len(email_list)):
    s = smtplib.SMTP('smtp.gmail.com', 587)
    s.starttls()
    s.login(gmail_user, "Sandeep@1997")
    body = '\r\n'.join(['send_to: %s' % email_list[i],
                        'From: %s' % gmail_user,
                        'Subject: %s' % SUBJECT,
                        '', message])
    s.sendmail(gmail_user, email_list[i], body)
    print(f"Mail sent to: {email_list[i]}")
    s.quit()
```
